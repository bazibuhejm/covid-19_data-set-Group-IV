```python
# Import dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
import numpy as np
#  Import and read the owid-covid-data.csv.
loc_df = pd.read_csv("loc_df.csv")
loc_df.head()
```

```python
# Drop the vaccination column from loc_df.
loc_df= loc_df.drop(['Unnamed: 0'],axis = 1)
loc_df.head()
```

```python
# Drop the vaccination column from loc_df.
loc_df= loc_df.drop(['location'],axis = 1)
loc_df.head()
```

```python
# Join every string in every row, split the result, pull out the unique values.
vaccine = np.unique(', '.join(loc_df['vaccines']).split(', '))
# Drop 'None'
vaccine = np.delete(vaccine, np.where(vaccine == 'None'))
```

```python
# One hot for loop...
for vax in vaccine:
    loc_df[vax] = loc_df['vaccines'].str.contains(vax).astype('int')

loc_df.drop('vaccines', axis=1, inplace=True)
loc_df
```

```python
loc_df.to_csv("vax_loc_onehot.csv")
```
